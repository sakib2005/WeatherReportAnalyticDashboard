# Data Visualization

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set style for plots
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

### Loading the data 
This will find the the cleaned air quality dataset 'AirQualityUCI_Cleaned.csv' and loads it into the Dataframe, assuming the dataset file is in the same directory.
If the file is not found, it will create a sample dataset to be used. 

```python
# Load and parse the air qualaity dataset
# Read the CSV
df = pd.read_csv('AirQualityUCI_cleaned.csv')
```

### DateTime
Create DateTime column to and setting it as index to transfrom the DataFrame into a time series.

```python
# Set DateTime as index
df = df.set_index('DateTime').sort_index()

# Display basic info
print("\n" + "="*60)
print("DATASET INFORMATION")
print("="*60)
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}\n")
print(f"Date range: {df.index.min()} to {df.index.max()}")
print(f"Total missing values: {df.isna().sum().sum()}")
```

# Visualizations

## Correlation analysis

```python
print("\n" + "="*60)
print("1. CORRELATION ANALYSIS")
print("="*60)

# Identify ground truth columns (ending with (GT))
gt_columns = [col for col in df.columns if '(GT)' in col and 'PT08' not in col]
print(f"Ground truth columns: {gt_columns}")

# Identify sensor columns (PT08 sensors)
sensor_columns = [col for col in df.columns if 'PT08' in col]
print(f"Sensor columns: {sensor_columns}")

# Combine for correlation analysis
analysis_columns = gt_columns + sensor_columns + ['T', 'RH', 'AH']

# Filter only columns that exist in the dataframe
analysis_columns = [col for col in analysis_columns if col in df.columns]

# Calculate correlation matrix
correlation_matrix = df[analysis_columns].corr()

# Separate correlations between sensors and ground truth
gt_sensor_corr = correlation_matrix.loc[gt_columns, sensor_columns]

print(f"\nCorrelation matrix shape: {correlation_matrix.shape}")
```

### 1. Complete Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, cbar_kws={'shrink': 0.8})
plt.title('Complete Correlation Matrix (All Variables)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 2. Sensors vs Ground Truth

```python
plt.figure(figsize=(8, 6))
sns.heatmap(gt_sensor_corr, annot=True, fmt='.2f', cmap='viridis', 
            center=0, square=True, cbar_kws={'shrink': 0.8})
plt.title('Correlation: Sensors vs Ground Truth', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 3. Top correlations with CO(GT) - target variable

```python
plt.figure(figsize=(10, 6))
# Get top 10 correlations with CO(GT)
target = 'CO(GT)'
if target in correlation_matrix.columns:
    corr_target = correlation_matrix[target].drop(target)
    top_corr = corr_target.abs().sort_values(ascending=False).head(10)
    top_corr_signed = corr_target[top_corr.index]
    
    colors = ['red' if x > 0 else 'blue' for x in top_corr_signed]
    top_corr_signed.plot(kind='barh', color=colors)
    plt.title(f'Top 10 Features Correlated with {target}', fontsize=14, fontweight='bold')
    plt.xlabel('Correlation Coefficient')
    plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```

### 4. Correlation network

```python
plt.figure(figsize=(10, 8))

# 1. Prepare Data
# Filter strong correlations (positive or negative)
threshold = 0.5
high_corr = correlation_matrix.stack().reset_index()
high_corr.columns = ['Var1', 'Var2', 'Correlation']
# Remove self-correlations (Var1 == Var2)
high_corr = high_corr[high_corr['Var1'] != high_corr['Var2']]
# Filter by threshold
high_corr = high_corr[high_corr['Correlation'].abs() > threshold]

# 2. Plotting Logic
if not high_corr.empty:
    # Get unique variables involved in high correlations
    unique_vars = pd.unique(high_corr[['Var1', 'Var2']].values.ravel('K'))
    
    # Assign positions in a CIRCLE (much cleaner than random)
    n_vars = len(unique_vars)
    pos = {}
    for i, var in enumerate(unique_vars):
        angle = 2 * math.pi * i / n_vars
        pos[var] = (math.cos(angle), math.sin(angle))
        
    # Draw connection lines
    for idx, row in high_corr.iterrows():
        x = [pos[row['Var1']][0], pos[row['Var2']][0]]
        y = [pos[row['Var1']][1], pos[row['Var2']][1]]
        
        # Color: Red for positive, Blue for negative
        color = 'red' if row['Correlation'] > 0 else 'blue'
        # Width/Alpha based on strength
        width = abs(row['Correlation']) * 3
        alpha = max(0.3, abs(row['Correlation']) - 0.1) 
        
        plt.plot(x, y, color=color, alpha=alpha, linewidth=width, zorder=1)
        
    # Draw points and labels
    for var, (x, y) in pos.items():
        # Draw a white circle behind text to make it readable
        plt.scatter(x, y, s=1000, color='white', edgecolors='gray', zorder=2)
        plt.text(x, y, var, fontsize=10, ha='center', va='center', fontweight='bold', zorder=3)

    plt.title(f'Correlation Network (|r| > {threshold})', fontsize=15, fontweight='bold')
    plt.axis('off') # Hide the square box/axis
    plt.tight_layout()
    plt.show()
else:
    print("No correlations found above threshold.")
```

## Correlation insights

```python
print("\n" + "="*60)
print("CORRELATION INSIGHTS")
print("="*60)

if 'CO(GT)' in correlation_matrix.columns:
    print("\nTop 5 positive correlations with CO(GT):")
    co_corr = correlation_matrix['CO(GT)'].sort_values(ascending=False)
    for idx, (var, corr) in enumerate(co_corr.head(6).items()):
        if var != 'CO(GT)':
            print(f"  {idx}. {var}: {corr:.3f}")
    
    print("\nTop 5 negative correlations with CO(GT):")
    for idx, (var, corr) in enumerate(co_corr.tail(6).items()):
        if var != 'CO(GT)':
            print(f"  {idx}. {var}: {corr:.3f}")
```

## Time Series Plots
Time series plots for selected month

```python
print("\n" + "="*60)
print("2. TIME-SERIES ANALYSIS")
print("="*60)

# Select a month for detailed analysis (choose a month with good data coverage)
# Let's select March 2004 (based on the data samples)
selected_month = '2004-03'
if selected_month in df.index.strftime('%Y-%m').unique():
    month_data = df[df.index.strftime('%Y-%m') == selected_month].copy()
else:
    # If selected month not available, pick first available month
    available_months = df.index.strftime('%Y-%m').unique()
    selected_month = available_months[0] if len(available_months) > 0 else '2004-03'
    month_data = df[df.index.strftime('%Y-%m') == selected_month].copy()

print(f"Selected month for time-series analysis: {selected_month}")
print(f"Number of observations in selected month: {len(month_data)}")

# Create time-series plots
fig, axes = plt.subplots(4, 2, figsize=(18, 16))
fig.suptitle(f'Time-Series Analysis - {selected_month}', fontsize=18, fontweight='bold', y=1.02)
```

### CO(GT) concentration


```python
if 'CO(GT)' in month_data.columns:
    axes[0, 0].plot(month_data.index, month_data['CO(GT)'], 
                    color='darkred', linewidth=1.5, alpha=0.8)
    axes[0, 0].fill_between(month_data.index, 0, month_data['CO(GT)'], 
                            alpha=0.3, color='darkred')
    axes[0, 0].set_title('CO(GT) Concentration', fontsize=12, fontweight='bold')
    axes[0, 0].set_ylabel('Concentration (mg/m)')
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].tick_params(axis='x', rotation=45)
    
    # Add statistics
    avg_co = month_data['CO(GT)'].mean()
    max_co = month_data['CO(GT)'].max()
    axes[0, 0].axhline(y=avg_co, color='black', linestyle='--', alpha=0.7, 
                       label=f'Avg: {avg_co:.2f}')
    axes[0, 0].legend(loc='upper right')
```

### 2. C6H6(GT) concentration

```python
if 'C6H6(GT)' in month_data.columns:
    axes[0, 1].plot(month_data.index, month_data['C6H6(GT)'], 
                    color='purple', linewidth=1.5, alpha=0.8)
    axes[0, 1].fill_between(month_data.index, 0, month_data['C6H6(GT)'], 
                            alpha=0.3, color='purple')
    axes[0, 1].set_title('Benzene (C6H6) Concentration', fontsize=12, fontweight='bold')
    axes[0, 1].set_ylabel('Concentration (g/m)')
    axes[0, 1].grid(True, alpha=0.3)
    axes[0, 1].tick_params(axis='x', rotation=45)
```

### 3. NOx(GT) concentration

```python
if 'NOx(GT)' in month_data.columns:
    axes[1, 0].plot(month_data.index, month_data['NOx(GT)'], 
                    color='darkblue', linewidth=1.5, alpha=0.8)
    axes[1, 0].fill_between(month_data.index, 0, month_data['NOx(GT)'], 
                            alpha=0.3, color='darkblue')
    axes[1, 0].set_title('NOx Concentration', fontsize=12, fontweight='bold')
    axes[1, 0].set_ylabel('Concentration (ppb)')
    axes[1, 0].grid(True, alpha=0.3)
    axes[1, 0].tick_params(axis='x', rotation=45)
```

### 4. NO2(GT) concentration

```python
if 'NO2(GT)' in month_data.columns:
    axes[1, 1].plot(month_data.index, month_data['NO2(GT)'], 
                    color='teal', linewidth=1.5, alpha=0.8)
    axes[1, 1].fill_between(month_data.index, 0, month_data['NO2(GT)'], 
                            alpha=0.3, color='teal')
    axes[1, 1].set_title('NO Concentration', fontsize=12, fontweight='bold')
    axes[1, 1].set_ylabel('Concentration (ppb)')
    axes[1, 1].grid(True, alpha=0.3)
    axes[1, 1].tick_params(axis='x', rotation=45)
```

### 5. Sensor PT08.S1(CO) readings

```python
if 'PT08.S1(CO)' in month_data.columns:
    axes[2, 0].plot(month_data.index, month_data['PT08.S1(CO)'], 
                    color='orange', linewidth=1.5, alpha=0.8)
    axes[2, 0].set_title('Sensor PT08.S1(CO) Readings', fontsize=12, fontweight='bold')
    axes[2, 0].set_ylabel('Sensor Reading')
    axes[2, 0].grid(True, alpha=0.3)
    axes[2, 0].tick_params(axis='x', rotation=45)
```

### 6. Environmental factors

```python
if all(col in month_data.columns for col in ['T', 'RH']):
    # Temperature
    ax_temp = axes[2, 1]
    color_temp = 'red'
    ax_temp.plot(month_data.index, month_data['T'], color=color_temp, 
                 linewidth=1.5, alpha=0.8, label='Temperature')
    ax_temp.set_ylabel('Temperature (C)', color=color_temp)
    ax_temp.tick_params(axis='y', labelcolor=color_temp)
    ax_temp.grid(True, alpha=0.3)
    
    # Relative Humidity (secondary axis)
    ax_rh = ax_temp.twinx()
    color_rh = 'blue'
    ax_rh.plot(month_data.index, month_data['RH'], color=color_rh, 
               linewidth=1.5, alpha=0.5, linestyle='--', label='Humidity')
    ax_rh.set_ylabel('Relative Humidity (%)', color=color_rh)
    ax_rh.tick_params(axis='y', labelcolor=color_rh)
    
    ax_temp.set_title('Environmental Factors', fontsize=12, fontweight='bold')
    ax_temp.tick_params(axis='x', rotation=45)
    
    # Combine legends
    lines_temp, labels_temp = ax_temp.get_legend_handles_labels()
    lines_rh, labels_rh = ax_rh.get_legend_handles_labels()
    ax_temp.legend(lines_temp + lines_rh, labels_temp + labels_rh, 
                   loc='upper right')
```

### 7. Daily patterns (aggregated by hour)

```python
if 'CO(GT)' in month_data.columns:
    daily_pattern = month_data.groupby(month_data.index.hour)['CO(GT)'].agg(['mean', 'std'])
    axes[3, 0].errorbar(daily_pattern.index, daily_pattern['mean'], 
                        yerr=daily_pattern['std'], 
                        fmt='o-', color='darkgreen', 
                        ecolor='lightgreen', elinewidth=2, capsize=4)
    axes[3, 0].set_title('Daily Pattern of CO(GT) (Hourly Average)', 
                         fontsize=12, fontweight='bold')
    axes[3, 0].set_xlabel('Hour of Day')
    axes[3, 0].set_ylabel('CO(GT) Concentration (mg/m)')
    axes[3, 0].grid(True, alpha=0.3)
    axes[3, 0].set_xticks(range(0, 24, 3))
```

### 9. Weekly patterns (if we have full weeks)

```python
if 'CO(GT)' in month_data.columns:
    month_data['Weekday'] = month_data.index.weekday
    weekly_pattern = month_data.groupby('Weekday')['CO(GT)'].agg(['mean', 'std'])
    weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    
    axes[3, 1].bar(weekly_pattern.index, weekly_pattern['mean'], 
                   yerr=weekly_pattern['std'], 
                   color=plt.cm.Set3(range(7)), 
                   edgecolor='black', linewidth=1, 
                   capsize=5, error_kw={'elinewidth': 2})
    axes[3, 1].set_title('Weekly Pattern of CO(GT)', fontsize=12, fontweight='bold')
    axes[3, 1].set_xlabel('Day of Week')
    axes[3, 1].set_ylabel('Average CO(GT) Concentration (mg/m)')
    axes[3, 1].set_xticks(range(7))
    axes[3, 1].set_xticklabels(weekdays)
    axes[3, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

## Monthly Statistics

```python
# Print monthly statistics
print(f"\nMonthly Statistics for {selected_month}:")
print("-"*40)

if 'CO(GT)' in month_data.columns:
    co_stats = month_data['CO(GT)'].describe()
    print(f"CO(GT) Statistics:")
    print(f"  Mean: {co_stats['mean']:.3f} mg/m")
    print(f"  Std:  {co_stats['std']:.3f} mg/m")
    print(f"  Min:  {co_stats['min']:.3f} mg/m")
    print(f"  Max:  {co_stats['max']:.3f} mg/m")
    print(f"  Missing: {month_data['CO(GT)'].isna().sum()} values")

# Create summary statistics table
print("\n" + "="*60)
print("SUMMARY STATISTICS")
print("="*60)

summary_stats = df[gt_columns].describe().T[['mean', 'std', 'min', 'max', 'count']]
summary_stats['missing'] = df[gt_columns].isna().sum().values
summary_stats['missing_pct'] = (summary_stats['missing'] / len(df) * 100).round(2)

print(summary_stats.to_string())

# Save cleaned data for next steps
df.to_csv('air_quality_cleaned.csv')
print("\n Cleaned data saved to 'air_quality_cleaned.csv'")
```
